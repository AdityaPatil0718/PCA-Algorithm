```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
dataset = load_iris()
x = dataset.data
y = dataset.target
```

```python
def pca(X, n_components):
    
    #convariance matrix
    cov_matrix = np.cov(x.T)

    #calculate eigenvalues and eigenvector
    eigenvalues, eigenvector = np.linalg.eig(cov_matrix)

    sorted_indices = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvector = eigenvector[:, sorted_indices]

    conponents = eigenvector[:, :n_components]
    projected_data = np.dot(x, n_components)

    return projected_data

n_components = 2
x_pca = pca(x, n_components)
print(x_pca)
```

```python
plt.figure(figsize=(8,6))
for i, target_name in enumerate(dataset.target_names):
    plt.scatter(x_pca[y == i,0], x_pca[y==i,1], label = target_name)

plt.title("pca for iris")
plt.legend()
plt.show()
```
